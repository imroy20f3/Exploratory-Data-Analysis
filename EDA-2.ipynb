```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Time_of_Day', y='money', hue='coffee_name')

# Move legend outside the plot
plt.legend(
    title='Coffee Name',
    bbox_to_anchor=(1.05, 1),   # position: (x, y)
    loc='upper left'             # anchor position
)

plt.title('Money by Time of Day and Coffee Type')
plt.tight_layout()  # adjust layout to fit everything
plt.show()
```

```python
df.groupby(['Month_name', 'coffee_name']).size().unstack().plot(kind='bar', stacked=True)
plt.legend(
    title='Coffee Name',
    bbox_to_anchor=(1.05, 1),   # position: (x, y)
    loc='upper left'             # anchor position
)

plt.tight_layout()  # adjust layout to fit everything
plt.show()
```

```python
#sns.lineplot(data=df, x='Date', y='money')
sample_df = df.sample(50, random_state=42)

sns.lineplot(data=sample_df, x='Date', y='money')
plt.title('Sample Trend of Money Over Date')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
